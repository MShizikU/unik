```python
import numpy as np
import statistics as st
import matplotlib.pyplot as plt

array = np.random.uniform(low=0, high=1, size=(10,))
average = st.mean(array)
median = st.median(array)

print("Average: ", average)
print("Median: ", median)
plt.scatter(range(0,10),array)
plt.ylabel('Array')
plt.show()
```

```python
import numpy as np
import math
import statistics as st
import matplotlib.pyplot as plt

array = []

for x in range(1,11):
    y = (math.sqrt(1 + math.e ** math.sqrt(x) + math.cos(x * x)))/(math.fabs(1 - math.sin(x) ** 3)) + math.log(2*x)
    array.append(y)
first_half = array[:math.ceil(len(array)/2)]
print(array)
print(first_half)

plt.scatter(range(0,math.ceil(len(array)/2)),first_half)
plt.ylabel('First half')
plt.show()

plt.plot(array)
plt.ylabel('Array')
plt.show()

```
